```python
def get_imdb_rating(title, api_key):
    try:
        # Reemplaza el espacio en blanco con '%20' para evitar errores en la URL
        title = title.replace(" ", "%20")
        
        # URL de la API con la clave y el título de la película
        url = f"http://www.omdbapi.com/?t={title}&apikey={api_key}"
        
        # Realizamos la solicitud
        response = requests.get(url)
        
        # Si la respuesta es válida
        if response.status_code == 200:
            data = response.json()
            
            if data['Response'] == 'True':
                return data.get('imdbRating', 'N/A')  # Devolver el imdbRating o 'N/A' si no se encuentra
            else:
                return f"No se encontró información para: {title}"
        else:
            return f"Error {response.status_code} en la búsqueda"
    
    except Exception as e:
        return f"Error al obtener datos: {str(e)}"
```
